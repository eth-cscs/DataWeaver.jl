```julia
using Pkg
Pkg.instantiate()
using IJulia
using LinearAlgebra
using ADIOS2
using GR
using Plots
using DataWeaver
```

```julia
init_data_weaver(filename = "adios2.xml", serial = true)     # specify the configuration file
read_setup("diffusion2D.bp")
```

```julia
step = 0
while reading()   
    DataWeaver.read("temperature")                                                       
    step += 1
    IJulia.clear_output(true)                                                  
    plot_title = "Temperature at step " * string(step)                   # Plot the temperature
    display(Plots.contourf(transpose(DataWeaver.V), title=plot_title, size=(700,500),
                     xtickfontsize=20, ytickfontsize=20,                 # ...
                     xguidefontsize=20, yguidefontsize=20,
                     legendfontsize=20, titlefontsize=30))               # ...
    
end
```
